```python
import pandas as pd
import matplotlib
import sqlite3

connection=sqlite3.connect('../data/checking-logs.sqlite')
```

```python
pd.read_sql(sql="SELECT datetime FROM pageviews WHERE uid LIKE 'user_%'", con=connection)
```

```python
df=pd.read_sql(sql="SELECT DATE(datetime) AS date, COUNT(uid) AS views FROM pageviews WHERE uid LIKE 'user_%' GROUP BY DATE(datetime);", con=connection)
df
```

```python
df.plot(x='date',y='views', figsize=(15, 8), title='Views per day', fontsize=8)
```

```python
connection.close()
```
