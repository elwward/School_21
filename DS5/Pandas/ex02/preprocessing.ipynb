```python
import pandas as pd
```

```python
df=pd.read_csv('../data/auto.csv', index_col='ID')
```

```python
df.count()
```

```python
df=df.drop_duplicates(keep='last', subset=['CarNumber', 'Make_n_model','Fines'])
```

```python
df.count()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=1,thresh=500)
df.isnull().sum()
```

```python
df['Refund'] = df['Refund'].fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
df['Fines'] = df['Fines'].fillna(value=df['Fines'].mean())
```

```python
df.isnull().sum()
```

```python
df['Make'] = df['Make_n_model'].apply(lambda x: x.split()[0])
df['Model'] = df['Make_n_model'].apply(lambda x: x.split()[1] if len(x.split())==2 else None)
```

```python
df=df.drop(columns='Make_n_model')
```

```python
df.to_json('../data/auto.json', orient='records')
```
