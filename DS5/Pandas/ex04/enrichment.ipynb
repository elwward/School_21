```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_json('../data/auto.json',orient='records')
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df
```

```python
sample = df[['CarNumber', 'Make', 'Model']].sample(n=200, replace=True, random_state=21).reset_index(drop=True)

sample['Refund'] = df['Refund'].sample(n=200, replace=True, random_state=21).reset_index(drop=True)

sample['Fines'] = df['Fines'].sample(n=200,replace=True, random_state=21).reset_index(drop=True)

sample
```

```python
concat_rows=pd.concat([df,sample])
concat_rows.index = range(len(concat_rows))
concat_rows
```

```python
np.random.seed(21)
years = pd.Series(np.random.randint(1980, 2019, size=925))
```

```python
fines=pd.concat([concat_rows,years],axis=1)
fines = fines.rename(columns={0: 'Year'})
fines  
```

```python
popular_surname=pd.read_json('../data/surname.json')
```

```python
popular_surname.columns = popular_surname.iloc[0] 
popular_surname = popular_surname[1:].reset_index(drop=True)
popular_surname
```

```python
k=fines['CarNumber'].nunique()
k
```

```python
surnames=popular_surname['NAME'].sample(n=k, replace=True, random_state=21).reset_index(drop=True)
surnames = surnames.str.replace(r'[^A-Za-z\s\-]', '', regex=True)
surnames
```

```python
cars=pd.Series(fines['CarNumber'].unique())
owners=pd.concat([cars,surnames], axis=1)

owners = owners.rename(columns={0: 'CarNumber'})
owners = owners.rename(columns={'NAME': 'SURNAME'})
owners
```

```python
new_rows = [
    ['FK7362JG322RUS', 1, 7200.00, 'Hyundai', 'Solaris', 2016],
    ['US738G728RUS', 1, 1800.00, 'Porsche', 'Cayenne', 2012],
    ['AA837YH192RUS', 1, 6500.00, 'Hyundai', 'Accent', 2010],
    ['GJ123H3P128RUS', 1, 7400.00, 'Lada', 'Granta', 2014],
    ['KP341H177RUS', 1, 9200.00, 'Ford', 'Focus', 2015]
]

temp = pd.DataFrame(new_rows, columns=fines.columns)
fines = pd.concat([fines, temp], ignore_index=True)

fines
```

```python
owners = owners[:-20]
owners
```
